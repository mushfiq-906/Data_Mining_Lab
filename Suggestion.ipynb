```python
import pandas as pd
import difflib

# Step 1: Read the first 100 rows of both CSV files
hapax_file_path = '/media/mushfiq/Academic/DM_LAB/filtered_hapax.csv'
unigram_file_path = '/media/mushfiq/Academic/DM_LAB/filtered_unigram.csv'

# Read the first 100 rows of the "filtered_hapax.csv" file into a DataFrame
hapax_data = pd.read_csv(hapax_file_path,nrows=50000)

# Read the first 100 rows of the "filtered_unigram.csv" file into a DataFrame
unigram_data = pd.read_csv(unigram_file_path,nrows=50000)

# Step 2: Tokenize the words from both DataFrames
hapax_words = set(hapax_data['processed_text'].str.split().sum())
unigram_words = set(unigram_data['processed_text'].str.split().sum())

# Step 3: Identify matched words
matched_words = hapax_words.intersection(unigram_words)

# Step 4: Extract unmatched words from the "filtered_hapax.csv" file
unmatched_words = hapax_words - matched_words

# Step 5: Provide suggestions for unmatched words
suggestions = {}
for unmatched_word in unmatched_words:
    closest_matches = difflib.get_close_matches(unmatched_word, unigram_words, n=3, cutoff=0.65)
    suggestions[unmatched_word] = closest_matches

# Define the paths to save the results in text files
matched_file_path = '/media/mushfiq/Academic/DM_LAB/matched_words.txt'  # Replace with your desired file path
unmatched_file_path = '/media/mushfiq/Academic/DM_LAB/unmatched_words.txt'  # Replace with your desired file path

# Save matched words to a text file
with open(matched_file_path, 'w', encoding='utf-8') as matched_file:
    matched_file.write("Matched Words:\n")
    matched_file.write('\n '.join(matched_words))

# Save unmatched words and suggestions to a text file
with open(unmatched_file_path, 'w', encoding='utf-8') as unmatched_file:
    unmatched_file.write("Unmatched Words and Suggestions:\n")
    for word, suggestion_list in suggestions.items():
        unmatched_file.write(f"{word}: Suggestions - {', '.join(suggestion_list)}\n")

# Print a message indicating where the results are saved
print(f"Matched words saved to {matched_file_path}")
print(f"Unmatched words and suggestions saved to {unmatched_file_path}")
```
